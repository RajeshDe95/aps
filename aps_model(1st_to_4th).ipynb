Creating Models

```python
data=pd.read_csv("imputed_median.csv")
```

```python
data.head()
```

```python
data.columns
```

```python
#Dropping the unwanted column 
data_tf=data.drop(columns=['Unnamed: 0'])
data_tf.head()
```

```python
# Segregating data into X & y 
X = data_tf.drop("class",axis=1)
y = data_tf['class']
```

```python
X
```

```python
y
```

```python
# Splitting data into train & test datasets
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
# Preprocessing
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X_train_tf=scaler.fit_transform(X_train)
```

```python
X_train_tf
```

```python
y
```

1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model_log=LogisticRegression()
```

```python
# Fitting the data train data into the model
model_log.fit(X_train_tf,y_train)
```

```python
## Trainng accuracy
model_log.score(X_train_tf,y_train) 
```

```python
#Transforming the Test data
X_test_tf=scaler.transform(X_test)
```

```python
y_pred_log=model_log.predict(X_test_tf)
```

```python
y_pred_log
```

```python
# importing library to check accuracy score
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_log)  
```

Observation For Logistic Regression
- Training Score of Logistic Regression: 99%
- Test Score of Logistic Regression: 99%

Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
conf_mat = confusion_matrix(y_test,y_pred_log)
conf_mat
```

```python
TP = conf_mat[0][0]
FP = conf_mat[0][1]
FN = conf_mat[1][0]
TN = conf_mat[1][1]
```

```python
Accuracy = (TP + TN) / (TP +FP + FN + TN)
Accuracy
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred_log,y_test))
```

3. Decesion Tree

```python
from sklearn.tree import DecisionTreeClassifier
model_dt=DecisionTreeClassifier()
```

```python
model_dt.fit(X_train,y_train)
```

```python
# model score
model_dt.score(X_train,y_train)
```

```python
y_pred_dt=model_dt.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred_dt)
```

```python
from sklearn import tree
fig=plt.figure(figsize=(25,15))
tree.plot_tree(model_dt,filled=True)
```

```python
fig.savefig("Decesion Tree on aps dataset.png")
```

```python
grid_param = {
    'criterion': ['gini', 'entropy'],
    'max_depth' : range(2,10,1),
    'min_samples_leaf' : range(1,8,1),
    'min_samples_split': range(2,8,1),
    'splitter' : ['best', 'random']
    
}
```

```python
from sklearn.model_selection import GridSearchCV
gridsearch=GridSearchCV(estimator=model_dt,param_grid=grid_param,cv=3,verbose=1)
```

```python
gridsearch.fit(X_train,y_train)
```

```python
gridsearch.best_params_
```

```python
model_with_best_params=DecisionTreeClassifier(criterion= 'gini',max_depth= 8,min_samples_leaf= 1,min_samples_split= 4,splitter= 'best')
```

```python
model_with_best_params.fit(X_train,y_train)
```

```python
y_pred_dt_hyper=model_with_best_params.predict(X_test)
accuracy_score(y_test,y_pred_dt_hyper)
```

Observation for Decesion Tree
- Before Hyperparameter Tuning accuracy : 98%
- After Hyperparameter Tuning accuracy : 99%
